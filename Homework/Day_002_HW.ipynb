#### 截取部分資料

```python
#找出男性 且AMT_INCOME_TOTAL前20名的使用者ID(SK_ID_CURR)、AMT_INCOME_TOTAL、AMT_CREDIT、AMT_ANNUITY等資料
result = app_train[app_train.CODE_GENDER == "M"]
result = result.sort_values(by='AMT_INCOME_TOTAL', ascending=False)[0:20]
result = result[result.columns[[0,7,8,9]]]
result.reset_index(drop=True)
```
