# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return (1 * (x > 0))


x = plt.linspace(-10,10,100)
plt.plot(x, ReLU(x), 'r')
plt.plot(x, dReLU(x), 'b')
plt.grid()

plt.text(0, 9, r'$f(x)= (abs(x) * (x > 0))$', fontsize=15)
plt.text(0, 2, r'$f(x)= (1 * (x > 0))$', fontsize=15)
plt.show()
```
